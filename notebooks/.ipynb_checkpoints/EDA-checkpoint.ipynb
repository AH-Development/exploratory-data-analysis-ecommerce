## ðŸ“Š Distribution des variables numÃ©riques

Cette Ã©tape nous aide Ã  comprendre les formes des distributions et repÃ©rer dâ€™Ã©ventuelles valeurs extrÃªmes (outliers).


```python
df.hist(bins=30, figsize=(12, 8), color='skyblue', edgecolor='black')
plt.tight_layout()
plt.show()
```

## ðŸ”— CorrÃ©lation entre les variables

On explore les relations linÃ©aires entre les variables Ã  lâ€™aide dâ€™une heatmap.


```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matrice de corrÃ©lation")
plt.show()
```

## âš ï¸ DÃ©tection des valeurs aberrantes

Nous utilisons des boxplots pour repÃ©rer les outliers dans certaines colonnes critiques.


```python
for col in ['Product_Price', 'Discount_Rate', 'Return_Rate']:
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f'Distribution de {col}')
    plt.show()
```

## ðŸ§¼ Traitement des valeurs manquantes

On applique diffÃ©rentes stratÃ©gies dâ€™imputation selon le type de donnÃ©es.


```python
df['Product_Rating'].fillna(df['Product_Rating'].median(), inplace=True)
df['Days_to_Deliver'].fillna(df['Days_to_Deliver'].median(), inplace=True)
```

## ðŸ” Suppression des doublons


```python
df.drop_duplicates(inplace=True)
```

## ðŸ”§ Correction des types de colonnes

Certaines colonnes doivent Ãªtre converties en types plus appropriÃ©s.


```python
df['Category_ID'] = df['Category_ID'].astype('category')
```

## ðŸš« Nettoyage des valeurs incohÃ©rentes


```python
df = df[df['Product_Price'] > 0]
df = df[df['Discount_Rate'].between(0, 1)]
df = df[df['Product_Rating'].between(0, 5)]
```

## ðŸ“ Normalisation des variables numÃ©riques

Pour prÃ©parer les donnÃ©es Ã  une modÃ©lisation ou Ã  une analyse plus avancÃ©e.


```python
from sklearn.preprocessing import StandardScaler

cols_to_scale = ['Product_Price', 'Discount_Rate', 'Days_to_Deliver', 'Return_Rate']
scaler = StandardScaler()
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
```
