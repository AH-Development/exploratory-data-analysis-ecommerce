# üßº Analyse Exploratoire et Nettoyage des Donn√©es

Ce notebook pr√©sente les √©tapes d'exploration et de nettoyage du dataset **`ecommerce_product_performance`**, en vue de le rendre exploitable pour des analyses ou des mod√®les pr√©dictifs.



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
%matplotlib inline
```

## üîπ Chargement des donn√©es
On commence par charger le dataset et afficher un aper√ßu.


```python
df = pd.read_csv("../data/ecommerce_product_performance.csv")
df.head()
```

On examine la structure des donn√©es, les types de colonnes, les valeurs manquantes et les statistiques descriptives.

## üìä Distribution des variables num√©riques

Cette √©tape nous aide √† comprendre les formes des distributions et rep√©rer d‚Äô√©ventuelles valeurs extr√™mes (outliers).


```python
df.hist(bins=30, figsize=(12, 8), color='skyblue', edgecolor='black')
plt.tight_layout()
plt.show()
```

## üîó Corr√©lation entre les variables

On explore les relations lin√©aires entre les variables √† l‚Äôaide d‚Äôune heatmap.


```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matrice de corr√©lation")
plt.show()
```

## ‚ö†Ô∏è D√©tection des valeurs aberrantes

Nous utilisons des boxplots pour rep√©rer les outliers dans certaines colonnes critiques.


```python
for col in ['Product_Price', 'Discount_Rate', 'Return_Rate']:
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f'Distribution de {col}')
    plt.show()
```

## üßº Traitement des valeurs manquantes

On applique diff√©rentes strat√©gies d‚Äôimputation selon le type de donn√©es.


```python
df['Product_Rating'].fillna(df['Product_Rating'].median(), inplace=True)
df['Days_to_Deliver'].fillna(df['Days_to_Deliver'].median(), inplace=True)
```

## üîÅ Suppression des doublons


```python
df.drop_duplicates(inplace=True)
```

## üîß Correction des types de colonnes

Certaines colonnes doivent √™tre converties en types plus appropri√©s.


```python
df['Category_ID'] = df['Category_ID'].astype('category')
```

## üö´ Nettoyage des valeurs incoh√©rentes


```python
df = df[df['Product_Price'] > 0]
df = df[df['Discount_Rate'].between(0, 1)]
df = df[df['Product_Rating'].between(0, 5)]
```

## üìè Normalisation des variables num√©riques

Pour pr√©parer les donn√©es √† une mod√©lisation ou √† une analyse plus avanc√©e.


```python
from sklearn.preprocessing import StandardScaler

cols_to_scale = ['Product_Price', 'Discount_Rate', 'Days_to_Deliver', 'Return_Rate']
scaler = StandardScaler()
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
```
